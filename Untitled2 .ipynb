### K-Means Clustering for Customer Segmentation

This section loads the provided sales data, performs K-Means clustering based on 'Revenue' and 'Quantity' to segment customers, visualizes these segments, and then exports the segmented data to a CSV file.

# New section

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the data using your exact verified file path
file_path = '/content/Thiranex_Sales_Data.xlsx (1).xlsx'
df = pd.read_excel(file_path)

print("--- Starting Automated Data Cleaning Workflow ---")

# 2. Automation: Handle Missing Values
for col in df.columns:
    if df[col].isnull().sum() > 0:
        if df[col].dtype == 'object':
            df[col].fillna('Unknown', inplace=True)
        else:
            df[col].fillna(df[col].median(), inplace=True)

# 3. Automation: Remove Duplicates
duplicate_count = df.duplicated().sum()
if duplicate_count > 0:
    df.drop_duplicates(inplace=True)
print(f"Automated Check: Removed {duplicate_count} duplicate rows.")

# 4. Automation: Standardize Strings & Format Dates
string_cols = df.select_dtypes(include=['object']).columns
for col in string_cols:
    df[col] = df[col].astype(str).str.strip()

if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])

# Export clean dataset
clean_file_name = 'Automated_Cleaned_Sales_Data.csv'
df.to_csv(clean_file_name, index=False)

print("--- Data Cleaning Finalized. Commencing Automated Report Generation ---")

# 5. Automation: Generate Visual Summaries
plt.figure(figsize=(12, 5))

# Subplot 1: Revenue by Category
plt.subplot(1, 2, 1)
sns.barplot(data=df, x='Category', y='Revenue', estimator=sum, errorbar=None, palette='Blues_r')
plt.title('Revenue Contribution by Product Category')
plt.xticks(rotation=15)

# Subplot 2: Order Distribution by Region
plt.subplot(1, 2, 2)
sns.countplot(data=df, x='Region', palette='Pastel1')
plt.title('Order Distribution Frequency Across Regions')

plt.tight_layout()
report_image_name = 'automated_report_summary.png'
plt.savefig(report_image_name, dpi=150)
plt.show()

# 6. Automation: Generate Text-Based KPI Audit Report
report_text_name = 'Automated_Data_Audit_Report.txt'
with open(report_text_name, 'w') as f:
    f.write("# THIRANEX AUTOMATED DATA AUDIT & KPI REPORT\n")
    f.write("==================================================\n\n")
    f.write(f"Total Rows Processed Successfully: {len(df)}\n")
    f.write(f"Duplicate Row Artifacts Cleared: {duplicate_count}\n\n")
    f.write("## METRIC KEY PERFORMANCE INDICATORS\n")
    f.write(f"- Total Aggregated Revenue Portfolio: INR {df['Revenue'].sum():,.2f}\n")
    f.write(f"- Absolute Operational Profit Volume: INR {df['Profit'].sum():,.2f}\n")
    f.write(f"- Cumulative Units Dispatched: {df['Quantity'].sum()} items\n")

print(f"Pipeline Finished! Download your files: {clean_file_name}, {report_image_name}, and {report_text_name}")
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load data and convert Date column to standard datetime
df['Date'] = pd.to_datetime(df['Date'])

# 2. Group revenue by Year and Month to build a historical time sequence
monthly_data = df.groupby(df['Date'].dt.to_period('M'))['Revenue'].sum().reset_index()
monthly_data['Month_Ordinal'] = np.arange(len(monthly_data)) + 1

# 3. Train a Linear Regression Model on historical data
X = monthly_data[['Month_Ordinal']]  # Features: Month index (1, 2, 3...)
y = monthly_data['Revenue']          # Target: Total Revenue

model = LinearRegression()
model.fit(X, y)

# 4. Forecast for the next 3 future months
future_months = np.array([[len(monthly_data) + 1],
                          [len(monthly_data) + 2],
                          [len(monthly_data) + 3]])
predictions = model.predict(future_months)

# Create a dataframe for the forecasted periods
last_date = df['Date'].max()
future_dates = [pd.to_datetime(last_date) + pd.DateOffset(months=i) for i in range(1, 4)]
forecast_df = pd.DataFrame({
    'Date': [d.strftime('%Y-%m') for d in future_dates],
    'Revenue': predictions,
    'Type': 'Forecasted'
})

historical_df = pd.DataFrame({
    'Date': monthly_data['Date'].dt.strftime('%Y-%m'),
    'Revenue': y,
    'Type': 'Historical'
})

# Combine historical and forecasted data for visualization
final_trend = pd.concat([historical_df, forecast_df], ignore_index=False)

# 5. Plot the Trend and Future Forecast
plt.figure(figsize=(10, 6))
sns.lineplot(data=historical_df, x='Date', y='Revenue', marker='o', label='Historical Revenue', linewidth=2.5)
sns.lineplot(data=forecast_df, x='Date', y='Revenue', marker='X', label='Forecasted Revenue (Prediction)', linestyle='--', color='red', linewidth=2.5)
plt.title('Revenue Trend Forecasting - Predictive Analytics Task 3')
plt.xlabel('Timeline (Months)')
plt.ylabel('Total Revenue')
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend()
plt.savefig('revenue_forecast_plot.png')
plt.show()

# Export predictions to CSV for submission
final_trend.to_csv('Task3_Predictive_Analytics_Results.csv', index=False)
print("Task 3 completed successfully!")
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load your data from the newly provided Excel file
# Note: Ensure the filename matches exactly what you uploaded
file_path = '/content/Thiranex_Sales_Data.xlsx (1).xlsx'
df = pd.read_excel(file_path)

print(f"Data loaded from: {file_path}")
print("First 15 rows of the data:")
display(df.head())
```

```python
# 2. Select Features for Clustering (Revenue and Quantity)
# Ensure these columns exist in your DataFrame
if 'Revenue' in df.columns and 'Quantity' in df.columns:
    X = df[['Revenue', 'Quantity']]

    # 3. Apply K-Means to create 3 segments
    # You can adjust n_clusters based on your analysis or domain knowledge
    kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42, n_init=10) # Added n_init for newer sklearn versions
    df['Segment'] = kmeans.fit_predict(X)

    print("K-Means clustering applied. First 5 rows with segment information:")
    display(df.head())

    # 4. Create the Visualization
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=df, x='Quantity', y='Revenue', hue='Segment', palette='viridis', s=100)
    plt.title('Customer Segmentation (K-Means Clustering)')
    plt.xlabel('Quantity')
    plt.ylabel('Revenue')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    # 5. Export for your GitHub submission
    output_filename = 'Task2_Segmented_Data.csv'
    df.to_csv(output_filename, index=False)
    print(f"Analysis complete! Download '{output_filename}' from the folder icon.")
else:
    print("Error: 'Revenue' or 'Quantity' columns not found in the DataFrame. Please check your data.")
```
